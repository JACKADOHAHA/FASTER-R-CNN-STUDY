```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
data_loss = pd.read_csv('./result/GoogLeNet-losses.csv').drop(['Unnamed: 0'], axis=1)
data_loss.rename(columns={'0': 'loss'}, inplace=True)
data_loss
```

```python
ax = sns.lineplot(data=data_loss, palette="husl", linewidth=2.5)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('GoogLeNet-losses')
fig= ax.get_figure()
fig.savefig('./result/GoogLeNet-losses.png', dpi = 300)
```

```python
data_acc = pd.read_csv('./result/GoogLeNet-eval_accuracy.csv')
data_acc
```

```python
data_acc = data_acc.T
data_acc = data_acc.values
df = pd.DataFrame(data_acc[1:,:], columns=data_acc[0,:])
df
```

```python
ax = sns.lineplot(data=df, palette="husl", linewidth=2)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('GoogLeNet-eval_accuracy')
fig= ax.get_figure()
fig.savefig('./result/GoogLeNet-eval_accuracy.png', dpi = 300)
```

```python
data_val_loss = pd.read_csv('./result/GoogLeNet-label.csv').drop(["Unnamed: 0"], axis=1)
data_val_loss
```

```python
ax = sns.violinplot(data=data_val_loss, palette="Set2")
plt.xlabel('label')
plt.title('GoogLeNet-label')
fig= ax.get_figure()
fig.savefig('./result/GoogLeNet-label.png', dpi = 300)
```

```python
data_pred = pd.read_csv('./result/GoogLeNet-predict.csv').drop(["Unnamed: 0"], axis=1)
data_pred
```

```python
ax = sns.boxplot(data=data_pred, palette="Set2")
plt.xlabel('epoch')
plt.ylabel('pred')
plt.title('GoogLeNet-predict')
fig= ax.get_figure()
fig.savefig('./result/GoogLeNet-predict.png', dpi = 300)
```
